```python
import sys
import csv
import os
import numpy as np
import scipy.io as sio
from scipy.spatial import distance

a = sio.loadmat('/media/thaovu/DATA/NCKH/NUS_WID_Low_Level_Features/test_ind.mat')
test_ind = a["M"]
test_ind = test_ind
a = sio.loadmat('/media/thaovu/DATA/NCKH/NUS_WID_Low_Level_Features/train_ind.mat')
train_ind = a["M"]
train_ind = train_ind

#a = sio.loadmat('/media/thaovu/DATA/NCKH/BIG DATA Multi features learning for image annotation/All_labels.mat')
#All_labels = a["All_labels"]
#X_labels = All_labels[train_ind,:]
#Y_labels = All_labels[test_ind,:]
X_labels = np.zeros((2400,12))
for i in range(12):
        X_labels[(i*200):((i*200)+200),i] = 1
        
Y_labels = np.zeros((1200,12))
for i in range(12):
        Y_labels[(i*100):((i*100)+100),i] = 1

########## top 100 BOW ##############
a = sio.loadmat('/media/thaovu/DATA/NCKH/NUS_WID_Low_Level_Features/BOW_int.mat')
BoW = a["BoW_int"]

Y_BOW = BoW[test_ind,:]
Y_BOW = Y_BOW.reshape(1200,500)
X_BOW = BoW[train_ind,:]
X_BOW = X_BOW.reshape(2400,500)

Dis = distance.cdist(Y_BOW, X_BOW, 'cityblock')

sort_ind = np.argsort(Dis)
BOW_index_top100 = sort_ind[:,0:100]

############### top 100 CH ############
a = sio.loadmat('/media/thaovu/DATA/NCKH/NUS_WID_Low_Level_Features/Nor_CH.mat')
CH = a["Normalized_CH"]

Y_CH = CH[test_ind,:]
Y_CH = Y_CH.reshape(1200,64)
X_CH = CH[train_ind,:]
X_CH = X_CH.reshape(2400,64)

Dis = distance.cdist(Y_CH, X_CH, 'cityblock')

sort_ind = np.argsort(Dis)
CH_index_top100 = sort_ind[:,0:100]

############### top 100 CM55 ############
a = sio.loadmat('/media/thaovu/DATA/NCKH/NUS_WID_Low_Level_Features/Nor_CM55.mat')
CM55 = a["Normalized_CM55"]

Y_CM55 = CM55[test_ind,:]
Y_CM55 = Y_CM55.reshape(1200,225)
X_CM55 = CM55[train_ind,:]
X_CM55 = X_CM55.reshape(2400,225)

Dis = distance.cdist(Y_CM55, X_CM55, 'cityblock')

sort_ind = np.argsort(Dis)
CM55_index_top100 = sort_ind[:,0:100]

############### top 100 CORR ############
a = sio.loadmat('/media/thaovu/DATA/NCKH/NUS_WID_Low_Level_Features/Nor_CORR.mat')
CORR = a["Normalized_CORR"]

Y_CORR = CORR[test_ind,:]
Y_CORR = Y_CORR.reshape(1200, 144)
X_CORR = CORR[train_ind,:]
X_CORR =X_CORR.reshape(2400, 144)

Dis = distance.cdist(Y_CORR, X_CORR, 'cityblock')

sort_ind = np.argsort(Dis)
CORR_index_top100 = sort_ind[:,0:100]

############### top 100 EDH ############
a = sio.loadmat('/media/thaovu/DATA/NCKH/NUS_WID_Low_Level_Features/Nor_EDH.mat')
EDH = a["Normalized_EDH"]

Y_EDH = EDH[test_ind,:]
Y_EDH = Y_EDH.reshape(1200, 73)
X_EDH = EDH[train_ind,:]
X_EDH = X_EDH.reshape(2400, 73)

Dis = distance.cdist(Y_EDH, X_EDH, 'cityblock')

sort_ind = np.argsort(Dis)
EDH_index_top100 = sort_ind[:,0:100]

############### top 100 WT ############
a = sio.loadmat('/media/thaovu/DATA/NCKH/NUS_WID_Low_Level_Features/Nor_WT.mat')
WT = a["Normalized_WT"]

Y_WT = WT[test_ind,:]
Y_WT = Y_WT.reshape(1200, 128)
X_WT = WT[train_ind,:]
X_WT = X_WT.reshape(2400, 128)

Dis = distance.cdist(Y_WT, X_WT, 'cityblock')

sort_ind = np.argsort(Dis)
WT_index_top100 = sort_ind[:,0:100]

################# gan nhan #############
n_class = 12
num_concept_per_imgs = 1
K_NN = 50
labels_predict = np.zeros(shape=(1,n_class))
for i in range(1200):
    BOW_100KNN_concepts = X_labels[BOW_index_top100[i,],]
    CH_100KNN_concepts = X_labels[CH_index_top100[i,],]
    CM55_100KNN_concepts = X_labels[CM55_index_top100[i,],]
    CORR_100KNN_concepts = X_labels[CORR_index_top100[i,],]
    EDH_100KNN_concepts = X_labels[EDH_index_top100[i,],]
    WT_100KNN_concepts = X_labels[WT_index_top100[i,],]

    img_labels = np.zeros(shape=(1,n_class))

    inter_joint_list = np.concatenate((BOW_100KNN_concepts[0:K_NN,], CH_100KNN_concepts[0:K_NN,], CM55_100KNN_concepts[0:K_NN,], CORR_100KNN_concepts[0:K_NN,], EDH_100KNN_concepts[0:K_NN,],WT_100KNN_concepts[0:K_NN,]), axis=0)
    inter_joint_list = inter_joint_list.reshape(300,12)

    out_labels = np.sum(inter_joint_list, axis = 0)
    ind_sort = np.argsort(out_labels)[::-1]
    result_index = ind_sort[0:num_concept_per_imgs]
    print (result_index)
    img_labels [0,result_index] = 1
    labels_predict = np.concatenate((labels_predict, img_labels),axis=0)
```
